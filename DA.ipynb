```python
features = ["range_ThermalConductivity", "wtd_std_ThermalConductivity", "range_atomic_radius", "wtd_gmean_ThermalConductivity",
            "std_ThermalConductivity", "wtd_entropy_Valence", "wtd_std_ElectronAffinity", "wtd_entropy_atomic_mass",
            "wtd_mean_Valence", "wtd_gmean_ElectronAffinity", "wtd_range_ElectronAffinity", "wtd_mean_ThermalConductivity",
            "wtd_gmean_Valence", "std_atomic_mass", "std_Density", "wtd_entropy_ThermalConductivity", "wtd_range_ThermalConductivity", 
            "wtd_mean_atomic_mass", "wtd_std_atomic_mass", "gmean_Density"]

for feature in features:
  plt.figure()
  sns.scatterplot(y=feature, x='critical_temp', data=data)
```

```python
for feature in features:
  plt.figure()
  sns.histplot(data[feature])
```

```python
for feature in features:
  plt.figure()
  sns.distplot(data[feature])
```
